```python
def plot_density(df,attributes):
    fig = plt.figure(figsize=(12, (len(attributes)/2)*2))
    for index, plot_vars in enumerate(attributes):
        ax = plt.subplot(len(attributes)/2, 3, index+1)
        final_y0 = final.loc[final['y'] == 0];
        final_y1 = final.loc[final['y'] == 1];
        ax = sns.kdeplot(data=final_y0[plot_vars], label='y=0', ax=ax)
        ax = sns.kdeplot(data=final_y1[plot_vars], label='y=1', ax=ax)
        a = plt.legend(['y=0','y=1']);
        ax.set(xlabel=plot_vars);
plt.tight_layout();
plt.show()
```

```python
def plot_box(df,attributes):
    fig = plt.figure(figsize=(12, (len(attributes)/2)*2))
    for index, plot_vars in enumerate(attributes):
        ax = plt.subplot(len(attributes)/2, 3, index+1)
        ax = sns.boxplot(y=final[plot_vars], x=final['y'])
        a = plt.legend(['y=0','y=1']);
        ax.set(xlabel=plot_vars);
    plt.tight_layout();
    plt.show()
    plot_density(final, nfeature)
    plot_box(final, nfeature)
```

```python
# Data Cleaning
# All features have missing values but less than 50, considering 160,000 rows data
# ['x24', 'x29', 'x30', 'x32']
# Fill missing values with mode:

obj_features = ['x24', 'x29', 'x30', 'x32']
for i in obj_features:
    final.loc[final[i].isna(), i] = final.loc[(~final[i].isna()), i].mode()[0]
    
# ['x37']
# Change feature 'x37' to numeric by removing '$' sign
final['x37'] = final['x37'].str.replace(r'$','')

# Change to numeric type
final['x37'] = final['x37'].astype(float)

# Fill missing values with mean
final.loc[final['x37'].isna(), 'x37'] = final.loc[(~final['x37'].isna()), 'x37'].mean()

# Fill numeric features with mean value
for i in list(nfeature):
    final.loc[final[i].isna(), i] = final.loc[(~final[i].isna()), i].mean()
    
final.info(verbose=True)
```

```python
# One-Hot Encoding on 4 object features: ['x24', 'x29', 'x30', 'x32']
obj_features = ['x24', 'x29', 'x30', 'x32']
for i in obj_features:
    tmp_df = pd.get_dummies(final[i], prefix = i)
    final = pd.concat((final, tmp_df), axis = 1)
    del final[i]
final.shape
final.info(verbose = True)
```
